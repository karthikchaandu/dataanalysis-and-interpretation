```
import pandas as pd

data = pd.read_csv('/content/gapminder.csv', low_memory=False)
 
print (len(data), "observations") #number of observations (rows)
print (len(data.columns), "variables") # number of variables (columns)
```

```
data['internetuserate'] = pd.to_numeric(data['internetuserate'], errors='coerce')
data['internetuserate']=pd.cut(data.internetuserate, 10)
 
data['incomeperperson'] = pd.to_numeric(data['incomeperperson'], errors='coerce')
data['incomeperperson']=pd.cut(data.incomeperperson, 10)
 
data['employrate'] = pd.to_numeric(data['employrate'], errors='coerce')
data['employrate']=pd.cut(data.employrate, 10)
 
data['femaleemployrate'] = pd.to_numeric(data['femaleemployrate'], errors='coerce')
data['femaleemployrate']=pd.cut(data.femaleemployrate, 10)
 
data['polityscore'] = pd.to_numeric(data['polityscore'], errors='coerce')
```

```
print ('counts for internetuserate – Percentage of people with access to internet')
c1 = data['internetuserate'].value_counts(sort=False)
print (c1)
print ('')
 
print ('percentages for internetuserate - Percentage of people with access to internet')
p1 = data['internetuserate'].value_counts(sort=False, normalize=True)
print (p1)
print ('')
 
print ('counts for incomeperperson - GDP per capita')
c2 = data['incomeperperson'].value_counts(sort=False)
print(c2)
print ('')
 
print ('percentages for incomeperperson - GDP per capita')
p2 = data['incomeperperson'].value_counts(sort=False, normalize=True)
print (p2)
print ('')
 
print ('counts for employrate – Percentage of people age 15+ who were employed in the year')
c3 = data['employrate'].value_counts(sort=False)
print(c3)
print ('')
 
print ('percentages for employrate - Percentage of people age 15+ who were employed in the year')
p3 = data['employrate'].value_counts(sort=False, normalize=True)
print (p3)
print ('')
 
print ('counts for femaleemployrate - Percentage of females age 15+ who were employed in the year')
c4 = data['femaleemployrate'].value_counts(sort=False)
print(c4)
print ('')
 
print ('percentages for femaleemployrate - Percentage of females age 15+ who were employed in the year')
p4 = data['femaleemployrate'].value_counts(sort=False, normalize=True)
print (p4)
print ('')
 
print ("counts for polityscore - The summary measure of a country's democratic and free nature. -10 is the lowest value, 10 the highest.")
c5 = data['polityscore'].value_counts(sort=False, bins=20)
print(c5)
print ('')
 
print ("percentages for polityscore - The summary measure of a country's democratic and free nature. -10 is the lowest value, 10 the highest.")
p5 = data['polityscore'].value_counts(sort=False, normalize=True, bins=20)
print (p5)
print ('')
```
